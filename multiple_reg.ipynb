```python
import pandas as pd
from sklearn import linear_model
df=pd.read_csv("C:\\Users\\Harsh Bhingarde\\Downloads\\cars.csv")
df.head()
```

```python
X=df[['Weight','Volume']].values
y=df[['CO2']].values
reg=linear_model.LinearRegression()
reg.fit(X,y)
predco2=reg.predict([[1150,1000]])
print(predco2)

predco2=reg.predict([[2300,1300]])
print(predco2)
```

```python
X
```

```python
print(reg.intercept_)
print(reg.coef_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
X1=df['Weight']
X2=df['Volume']
x1_range=np.linspace(X1.min(),X1.max(),100)
x2_range=np.linspace(X2.min(),X2.max(),100)

x1_mesh,x2_mesh=np.meshgrid(x1_range,x2_range)

x_mesh=np.c_[x1_mesh.ravel(),x2_mesh.ravel()]
y_pred_mesh=reg.predict(x_mesh)

y_pred_mesh=y_pred_mesh.reshape(x1_mesh.shape)

fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')

ax.scatter(X1,X2,y,color='blue',label="actual data")
ax.plt_surface(x1_mesh,x2_mesh,y_pred_mesh,color='red',alpha=0.5)
plt.show()

```
