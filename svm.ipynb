```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron,LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
from sklearn import metrics
```

```python
bc=datasets.load_breast_cancer()
bc
```

```python
X=bc.data
y=bc.target
print(X.shape)
print(y.shape)
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.3,random_state=1,stratify=y)
```

```python
sc=StandardScaler()
sc.fit(X_train)
X_train_std=sc.transform(X_train)
X_test_std=sc.transform(X_test)
```

```python
svc=SVC(C=1.0,random_state=1,kernel='linear')
svc.fit(X_train_std,Y_train)
```

```python
from sklearn.metrics import accuracy_score
#accuracy=accuracy_score()
y_predict=svc.predict(X_test_std)
print("accuracy score ",accuracy_score(Y_test,y_predict))
```

```python
from sklearn.inspection import DecisionBoundaryDisplay

# importing scikit learn with make_blobs
from sklearn.datasets import make_blobs

# creating datasets X containing n_samples
# Y containing two classes
X, y = make_blobs(n_samples=100, centers=2,random_state=0, cluster_std=0.40)
print(y.shape)
print(y.shape)

# plotting scatters
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='spring');
plt.show()
```

```python
# creating linspace between -1 to 3.5
xfit = np.linspace(-1, 3.5)
# plotting scatter
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='spring')
# plot a line between the different sets of data
for m, b, d in [(1, 0.65, 0.33), (0.5, 1.6, 0.55), (-0.2, 2.9, 0.2)]:
    yfit = m * xfit + b
    plt.plot(xfit, yfit, '-k')
    plt.fill_between(xfit, yfit - d, yfit + d, edgecolor='none',
    color='#AAAAAA', alpha=0.4)

plt.xlim(-1, 3.5);
plt.show()
```

```python
# fit the model, don't regularize for illustration purposes
# clf = svm.SVC(kernel="linear", C=1000)
# clf.fit(X, y)
from sklearn import svm


C = 1.0  # SVM regularization parameter
### SVC with linear kernel
# clf = svm.SVC(kernel='linear', C=C).fit(X, y)

###LinearSVC (linear kernel)
# clf= svm.LinearSVC(C=C).fit(X, y)

###SVC with RBF kernel
clf= svm.SVC(kernel='rbf', gamma=0.7, C=C).fit(X, y)

###SVC with polynomial (degree 3) kernel
# clf= svm.SVC(kernel='poly', degree=3, C=C).fit(X, y)



plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

# plot the decision function
ax = plt.gca()
DecisionBoundaryDisplay.from_estimator(
    clf,
    X,
    plot_method="contour",
    colors="k",
    levels=[-1, 0, 1],
    alpha=0.5,
    linestyles=["--", "-", "--"],
    ax=ax,
)
# plot support vectors
ax.scatter(
    clf.support_vectors_[:, 0],
    clf.support_vectors_[:, 1],
    s=100,
    linewidth=1,
    facecolors="none",
    edgecolors="k",
)
plt.show()
```
